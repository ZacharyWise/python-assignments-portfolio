```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point, max_steps=10000, tolerance=1e-10):
    def deriv(f, x):
        h = 1e-10
        return (f(x + h) - f(x - h)) / (2 * h)

    x_coords = [initial_point]
    y_coords = [f(initial_point)]

    for _ in range(max_steps): #Initiates a loop that runs for a maximum number of steps
        grad = deriv(f, x_coords[-1]) #Sets the deriv function to a variable
        if abs(grad) < tolerance: #Checks if the tolerance is above the absolute value of a variable, and if it is breaks the loop
            break
        next_x = x_coords[-1] - learning_rate * grad #Updates x by an amount equal to the learning steps value
        x_coords.append(next_x) #Appends the x_coords list 
        y_coords.append(f(next_x))  #Appends the x_coords list

    # Plotting
    plot_range = np.linspace(min(x_coords) - 1, max(x_coords) + 1, 1000)
    plt.plot(plot_range, [f(x) for x in plot_range], label='Function', color='blue')
    plt.plot(x_coords, y_coords, color='orange', linestyle='-', label='Descent Path')  # Orange line, no dots
    plt.legend()
    plt.title("Gradient Descent Path")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)
    plt.show()

    return round(x_coords[-1], 3), round(y_coords[-1], 3)

# Example usage
f = lambda x: x  # Define the function to minimize
gradient_descent(f, learning_rate=0.8, initial_point=1)  
```
