```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# 4a: Compute first partial derivatives of f(x, y) = e^x * sin(y) + y^3
x, y = sp.symbols('x y')
f = sp.exp(x) * sp.sin(y) + y**3

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
print("First partial derivatives:")
print("∂f/∂x:", df_dx)
print("∂f/∂y:", df_dy)

# 4b: Compute gradient vector and magnitude for g(x, y) = x^2y + xy^2 at (1, -1)
g = x**2 * y + x * y**2
grad_g = [sp.diff(g, x), sp.diff(g, y)]
magnitude = sp.sqrt(grad_g[0]**2 + grad_g[1]**2).subs({x: 1, y: -1})
print("\nGradient vector:")
print("∇g:", grad_g)
print("Magnitude at (1, -1):", magnitude)

# 4c: Compute second partial derivatives of h(x, y) = ln(x^2 + y^2)
h = sp.ln(x**2 + y**2)
h_xx = sp.diff(h, x, x)
h_yy = sp.diff(h, y, y)
h_xy = sp.diff(h, x, y)
print("\nSecond partial derivatives:")
print("∂²h/∂x²:", h_xx)
print("∂²h/∂y²:", h_yy)
print("∂²h/∂x∂y:", h_xy)
print("Symmetry check (∂²h/∂x∂y == ∂²h/∂y∂x):", h_xy == sp.diff(h, y, x))

# 4d: Contour plot for j(x, y) = x^3 - 3xy + y^3
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# 4e: Contour plot for a chosen function k(x, y) = sin(x) * cos(y)
k = sp.sin(x) * sp.cos(y)
k_func = sp.lambdify((x, y), k, 'numpy')
Z_k = k_func(X, Y)
plt.contourf(X, Y, Z_k, levels=50, cmap='plasma')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = sin(x) \cdot cos(y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
